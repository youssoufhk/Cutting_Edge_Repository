#**We begin by importing the libraries**


```


import pandas
import numpy as np
from matplotlib import font_manager as fm, rcParams
import matplotlib.pyplot as plt
import os


```

#Just to make the plot look fancy you can use this font 


```

fpath = os.path.join(rcParams["datapath"], "fonts/ttf/cmr10.ttf") 
prop = fm.FontProperties(fname=fpath)
fname = os.path.split(fpath)[1]

```

#To read the data from the Excel .csv file


```

d = pandas.read_csv('Book1.csv', delimiter=';')
data = d
pandas.DataFrame.head(data) #this shows you the first lines of your data


```

#Stock your variables


```

t = data['TIME']
x1 = data['Nee']
x2 = data['Nth']
x3 = data['PV/T']
x4 = data['E']

```

# Some parametrisation followed by 
# Proper plot

```

fig= plt.figure(figsize=(16,9)) #as you guessed it is the size
rcParams.update({'figure.autolayout': True})
#when the size is big it can cut the text so this is your savior



ax1= fig.add_subplot(111) #this is to declare the first axis (for x1,2,3)

# this is your time variable. since it is string you can't plot t directly
u = [i for i in range(len(t))] 

# you want the t apearing as Day Time for that use this
plt.xticks(u,t, rotation = 40, fontsize = 14) 

# instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()  





p1, = ax1.plot(u, x1, 'o-', label="Nee")
p2, = ax1.plot(u, x2, 's-', label="Nth")
p3, = ax1.plot(u, x3, '--', label="PV/T")
p4, = ax2.plot(u, x4, 'rd-', label="E")

ax1.set_xlabel(r'Day Time',fontproperties=prop, fontsize = 53)
ax1.set_ylabel(r'Efficiency %',fontproperties=prop, fontsize = 53, color=p1.get_color())
ax1.tick_params(axis='y', labelcolor=p1.get_color(), labelsize = 13)

ax2.set_ylabel(r'Irradiation $W m^{-2}$',fontproperties=prop, fontsize = 53, color=p4.get_color())  
# we already handled the x-label with ax1

ax2.tick_params(axis='y', labelcolor=p4.get_color(), labelsize = 13)

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()


# now for the legend and the fancy fonts etc.
font = fm.FontProperties(fname=fpath, weight='bold', style='normal', size=20)
ax2.legend(lines + lines2, labels + labels2, loc=0, prop=font)

plt.draw()
plt.show()
```
